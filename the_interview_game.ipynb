```python
"""
Docstring:

Introduction:
Thus game mocks the interview process for candidates applying for job in the field of Data Science / 
Data Analytics / Data Engineering. A virtual computer asks a few basic questions, the answers to which might make 
the candidate advance forward and get the job or loose out.

Round 1: Interview with HR
Round 2: Interview with Reporting Manager (Round 1)
Round 3: Interview with Reporting Manager (Round 2)

"""
# Importing necessary packages
from sys import exit

# Declaring global variables
def entrance_door():
    global candidate_first_name
    global candidate_last_name
    global position_applied_for
    
    
    print(f"Welcome to Datatronics. I am Suzan, here to help you with the sign-in.\n")
    input("<Press Enter key to proceed>\n")
    print("""\nCongrats on being shortlisted for the interview!!! I will need to confirm a few details before i let 
you go through this door.\n""")
    
    candidate_last_name = input("What is your last name? \t>") 
    candidate_first_name = input("What is your first name? \t>")
    position_applied_for = input("For what position are you interviewing today? \t>")
    
    print(f"""\nWelcome to Datatronics {candidate_first_name} {candidate_last_name}. We are extremely delighted to 
have you here. Good luck on your interview for the {position_applied_for} role.\n""")
    
    input("<Press Enter key to proceed. Then proceed to the hr round.>\n")
    
    hr_round_1()
    
# Defining functions    
def hr_round_1():
    print(f"\nHi {candidate_first_name}! My name is Ann and i am the HR manager at Datatronics.\n")
    input("<Press Enter key to proceed>\n")
    print("\nThis round is to get to know you a little better, before you proceed to your technical interviews.\n")
    input("<When you are ready, press Enter key>")
    print("""\nWill you now or in the future require Datatronics to sponsor you a visa for continuing your stay in
The United States?
1) no
2) yes
3) don't know\n""")
    
    choice = input("> ")
    choice = choice.lower()
    
    if "1" in choice or "no" in choice:
        print("\nAlright. That sounds good.\n")
        input("<Press Enter key to proceed>")
        hr_round_2()
        
    elif "2" in choice or "yes" in choice:
        print("\nI am sorry to inform but we do not sponsor a visa for this role.\n")
        input("<Press Enter key to proceed>")
        fail()
        
    elif "3" in choice or "don't know" in choice or "don't" in choice or "know" in choice:
        print("\nOk. This might need a glance from your Reporting Manager. Please clarify your status with him.\n")
        input("<Press Enter key to proceed>")
        hr_round_2()
    
    else:
        print("Didn't get that. Please select from the options and try again.")
        input("<Press Enter key to continue>\n")
        hr_round_1()

# Defining functions
def hr_round_2():
    print("""\nWhat are your salary expectations per annum?""")
    figure = input("Input digits after US$ like 12345 without commas.\t>US$ ")
    
    while True:
        
        try:
            # Converting figure into an integer for comparisons
            figure = int(figure)
            break
            
        except:
            print("\nPlease follow the instructions and input a valid number.\n")
            figure = input("\nInput digits after US$ like 12345 without commas.\t>US$ ")
    
    if figure <= 90000:
        print("\nAlright. That sounds good. Good luck for the next round with your Reporting Manager.\n")
        input("<Press Enter key to proceed>")
        reporting_manager_round_1()
        
    elif figure > 90000:
        print("\nThis might need a glance from your Reporting Manager. Good luck for the next round.\n")
        input("<Press Enter key to proceed>")
        reporting_manager_round_1()
        
    else:
        print("\nDidn't get that. Please input a valid number as directed.\n")
        hr_round_2()

# Defining functions
def reporting_manager_round_1():
    print(f"""\nHi {candidate_first_name}! I am Bob, your Reporting Manager. I would like to make sure that you are 
a perfect fit for this job.\n""")
    input("<When you are ready, press Enter key>\n")
    print("We are loooking to hire individuals with the right amount of experience for this role.")

# Declaring objects
    desired_experience = 4
    experience = 4

# Looping until candidate experience is a good match. ###There is a bug as the loop does not execute properly.
    while experience > 0:
        candidate_experience = input("\nHow many years of relevant experience do you have?\t>")
        
        while True:
            
            try:
                #converting candidate_experience for numerical comparison
                candidate_experience = int(candidate_experience)
                break
                
            except:
                print("\nSorry, that is a wrong input\n")
                candidate_experience = input("\nHow many years of relevant experience do you have?\t>")
                
        
        if candidate_experience == desired_experience:
            print("\nThat's great. You look like a good fit.\n")
            reporting_manager_round_2()
            
        elif candidate_experience != desired_experience:
            experience -= 1
            
            if experience > 0 and experience < 4:
                print("\nWe would like to test your skills further.\n")
                reporting_manager_round_2()
            
            elif experience == 0:
                fail()
                
        else:
            print(f"\n{candidate_first_name}, you seem over-qualified for this role.\n")
            choose = input("\nWould you like to proceed further? (Yes/No)\t>")
            choose = choose.lower()
            
            if choose == 'yes':
                reporting_manager_round_2()
                
            else:
                print(f"""\n{candidate_first_name}, thanks for considering Datatronics. Do check out our career 
website for regular updates on new jobes.\n""")
                exit

# Defining functions
def reporting_manager_round_2():
    print("\nThis role also requires a person with great coding skills.\n")

# Declaring objects
    desired_coding_experience = 3
    coding = 3

# Looping until candidate coding experience is close to desired. ### There is a bug as the loop does not end right.
    while coding > 0:
        candidate_coding_experience = input("\nHow many languages can you code in?\t>")
        
        while True:
            
            try:
                #converting candidate_experience for numerical comparison
                candidate_coding_experience = int(candidate_coding_experience)
                break
                
            except:
                print("\nSorry, that is a wrong input\n")
                candidate_coding_experience = input("\nHow many languages can you code in?\t>")
        
        if candidate_coding_experience == desired_coding_experience:
            print(f"\nCongrats {candidate_first_name}! You meet all the requirements for this job. Welcome on board!")
            break
            
        elif candidate_coding_experience != desired_coding_experience:
            coding -= 1
            
            if coding > 0 and coding < 3:
                print("""\nYou look like a promising candidate. Although we will keep you on hold right now and 
update you in a week's time.\n""")
                exit(0)
                
            elif coding == 0:
                fail()
            
# Defing functions
def fail():
    print(f"""\nWe are sorry {candidate_first_name}, but you don't seem to meet all the criteria required for this
role. Good luck with your job search.\n""")
    print("\nWould you like to be considered for our internship roles? \n")
    
    choice = input("Please type yes or no.\t>")
    choice = choice.lower()
    
    if choice == 'yes':
        print("\nWe will keep your cv on record and inform you of the internship opppurtunities at Datatronics\n")
        exit(0)
        
    elif choice == 'no':
        print(f"""{candidate_first_name}, thanks for your valuable time. We wish you all the very best in  your 
future endeavors!""")
        exit(0)
    
    
entrance_door()
```
